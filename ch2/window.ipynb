```python
for w in p.getWindowsWithTitle("제목 없음"):
    print(w)    
```

```python
w = p.getWindowsWithTitle("제목 없음")
if not w.isActive:
    w.activate() # 창 활성화

if not w.isMaximized:
    w.maximize()

if not w.isMinimized:
    w.minimize()
w.restore() # 원래 크기로
```

```python
import pyperclip
w = p.getWindowsWithTitle("제목 없음")
if not w.isActive:
    w.activate() # 창 활성화

p.write("write")
p.write("pyautogui",interval=0.25)
# p.write("안녕하세요") 한글작성 안됨

pyperclip.copy("안녕하세요")
p.hotkey("ctrl","v")
```

```python
def word_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
if not w.isActive:
    w.activate()

word_write("writer")
word_write("반갑습니다")
```
