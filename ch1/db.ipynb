```python
import random
import time
```

```python
# word.txt 단어 읽은 후 리스트에 추가하기
words = []
with open("./resource/word.txt","r") as f:
    for w in f:
        # print(w,end="")
        words.append(w.strip())
words
```

```python
input("Ready? press enter key")
start = time.time() # 시작 시간
n = 1
n,cor_cnt = 1, 0 # 시도횟수, 정답개수

while n <= 5 :
    # 리스트 요소 임의로 섞기
    random.shuffle(words)
    # 무작위로 하나만 추출
    q = random.choice(words)
    # 추출된 요소 보여주기
    print(f"Question #{n}")
    print(q)
    # 사용자로 부터 타이핑 입력받기
    answer = input()
    # 추출된 요소 == 타이핑 값(정확히 타자 친 개수에 추가)
    if str(q).strip() == str(answer).strip():
        print("Pass!!")
        cor_cnt += 1 # 정답 개수 추가
    else:
        print("Wrong!!")
    n += 1 # 문제 개수 추가

# 끝난시간
end = time.time()
# 걸린시간과 정답개수 출력
et = end - start
print("게임시간 : {}초, 정답개수 : {}".format(et,cor_cnt))

# 정답개수가 4개 이상인 경우 합격, 불합격
if cor_cnt >= 4:
    print("합격")
else:
    print("불합격")

# 테이블에 저장
conn = sqlite3.connect("./database/test.db",isolation_level = None)
# records
cursor = conn.cursor()

cursor.execute("CREATE TABLE IF NOT EXISTS records(cor_cnt integer, record text, regdate text)")

# 정답개수, 걸린시간, 날짜
# now = datetime.now()
now_date_time = now.strftime("%Y-%m-$d %H:%M:%S")
cursor.execute("INSERT INTO records(cor_cnt, record, regdate) VALUES(?,?,?)",(cor_cnt,et,now_date_time))
```
